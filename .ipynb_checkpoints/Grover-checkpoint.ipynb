# Grover Algorithmus

```python
#initialization
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg' # Makes the images look nice

# importing Qiskit
from qiskit import IBMQ, Aer
from qiskit.providers.ibmq import least_busy
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, execute

# import basic plot tools
from qiskit.visualization import plot_histogram
```

Als erstes schauen wir uns den Grover Algorithmus für N=4 an, was mit 2 Qubits realisiert werden kann. Für diesen Fall benötigen wir nur eine Rotation, welche den Anfangzustand in den Endzustand rotiert.

Hierfür gibt es 4 verschiedene Oracle, welche den Endzustand repräsentieren können. Diese Endzustände bei N=4 möglichen Elementen wären |00⟩,|01⟩,|10⟩,|11⟩.

### Oracle für |w⟩=|11⟩
Um ein sign flip für |11⟩ zu realisieren, müssen wir einfach ein controlled Z gate  auf den Anfangszustand anwenden. Dies führt zu dem folgenden Circuit:

```python
def three(circuit):
    circuit.cz(0, 1)
```

### Oracle für |w⟩=|00⟩
Um die sign flip für |00⟩ zu realisieren, müssen wir ein "inverted" controlled Z gate auf den Anfangszustand anwenden. Dies führt zu dem folgenden Circuit:

```python
def null(circuit):
    for qubit in range(2):
        circuit.x(qubit)

    circuit.cz(0, 1)

    for qubit in range(2):
        circuit.x(qubit)
```

### Oracles für |w⟩=|01⟩

```python
def one(circuit):
    circuit.x(1)
    circuit.cz(0, 1)
    circuit.x(1)
```

### Oracles für |w⟩=|10⟩

```python
def two(circuit):
    circuit.x(0)
    circuit.cz(0, 1)
    circuit.x(0)
```

#### Visualisierung der Oracles 

```python
circuit = QuantumCircuit(2)

## init
for qubit in range(2):
    circuit.h(qubit)

three(circuit)

circuit.draw('mpl')
```

Für dieses Beispiel wollen wir den Zustand |00⟩ erhalten. Hierfür benötigen wir noch eine letzte Operation

## Reflection Us
Um den circuit fertigzustellen, müssen wir zusätzlich die Reflection Us implementieren, welche folgende Eigenschaft hat

Us12(|00⟩+|01⟩+|10⟩+|11⟩)=12(|00⟩−|01⟩−|10⟩−|11⟩)

Dies hat zur Folge, dass alles Zustände geflippt werden außer |00⟩. Das zugehörige Oracle dafür:

```python
def reflection(circuit):
    for qubit in range(2):
        circuit.z(qubit)
    
    circuit.cz(0, 1)
```

## Anwendung

```python
n = 2
grover_circuit = QuantumCircuit(n)

# init
for qubit in range(n):
    grover_circuit.h(qubit)

grover_circuit.barrier()
    
# Oracle
one(grover_circuit)

grover_circuit.barrier()

# Post Oracle
for qubit in range(n):
    grover_circuit.h(qubit)
    
grover_circuit.barrier()
    
# reflection
reflection(grover_circuit)

grover_circuit.barrier()

# Post Reflection
for qubit in range(n):
    grover_circuit.h(qubit)
    
grover_circuit.draw('mpl')
```

```python
grover_circuit.measure_all()

backend = Aer.get_backend('qasm_simulator')
shots = 1024
results = execute(grover_circuit, backend=backend, shots=shots).result()
answer = results.get_counts()
plot_histogram(answer)
```

## Grover for 3 Qubits

Oracle for |110⟩

Um dieses Oracle zu implementieren, brauchen wir ein controlled controlled z gate. Dieses ist eine Kombination aus verschiedenen Gates, nämlich dem Hadamard und dem Toffoli gate (auch controlled controlled not gate genannt).

```python
ccz = QuantumCircuit(3)
ccz.h(2)
ccz.ccx(0,1,2)
ccz.h(2)

ccz.draw('mpl')
```

```python
def oracle(circuit):
    circuit.x(0)
    
    # CCZ
    circuit.h(2)
    circuit.ccx(0,1,2)
    circuit.h(2)
    
    circuit.x(0)
```

```python
def diffuser(circuit):
    for q in range(3):
        circuit.h(q)
        circuit.x(q)
    
    # CCZ
    circuit.h(2)
    circuit.ccx(0,1,2)
    circuit.h(2)
    
    for q in range(3):
        circuit.x(q)
        circuit.h(q)
```

```python
grover_circuit = QuantumCircuit(3)

for qubit in range(3):
    grover_circuit.h(qubit)

grover_circuit.barrier()

oracle(grover_circuit)

grover_circuit.barrier()

diffuser(grover_circuit)

grover_circuit.draw('mpl')
```

```python
grover_circuit.measure_all()

backend = Aer.get_backend('qasm_simulator')
shots = 1024
results = execute(grover_circuit, backend=backend, shots=shots).result()
answer = results.get_counts()
plot_histogram(answer)
```

## Run on Real Quantum Computer

```python
# Load IBM Q account and get the least busy backend device
provider = IBMQ.load_account()
device = least_busy(provider.backends(filters=lambda x: x.configuration().n_qubits >= 3 and 
                                   not x.configuration().simulator and x.status().operational==True))
print("Least busy device: ", device)
```

```python
# Run our circuit on the least busy backend. Monitor the execution of the job in the queue
from qiskit.tools.monitor import job_monitor
job = execute(grover_circuit, backend=device, shots=1024, max_credits=10)
job_monitor(job, interval = 2)
```

```python
# Get the results from the computation
results = job.result()
answer = results.get_counts(grover_circuit)
plot_histogram(answer)
```
